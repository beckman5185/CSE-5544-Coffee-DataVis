```python
# Note: Edit the path here to match your setup.
data_df = pd.read_csv('../coffee_survey.csv')
data_df.drop('submission_id', axis=1, inplace=True)
```

```python
data_df.drop('coffee_a_bitterness', axis=1, inplace=True)
data_df.drop('coffee_a_acidity', axis=1, inplace=True)
data_df.drop('coffee_a_personal_preference', axis=1, inplace=True)
data_df.drop('coffee_a_notes', axis=1, inplace=True)
data_df.drop('coffee_b_bitterness', axis=1, inplace=True)
data_df.drop('coffee_b_acidity', axis=1, inplace=True)
data_df.drop('coffee_b_personal_preference', axis=1, inplace=True)
data_df.drop('coffee_b_notes', axis=1, inplace=True)
data_df.drop('coffee_c_bitterness', axis=1, inplace=True)
data_df.drop('coffee_c_acidity', axis=1, inplace=True)
data_df.drop('coffee_c_personal_preference', axis=1, inplace=True)
data_df.drop('coffee_c_notes', axis=1, inplace=True)
data_df.drop('coffee_d_bitterness', axis=1, inplace=True)
data_df.drop('coffee_d_acidity', axis=1, inplace=True)
data_df.drop('coffee_d_personal_preference', axis=1, inplace=True)
data_df.drop('coffee_d_notes', axis=1, inplace=True)
data_df.drop('prefer_abc', axis=1, inplace=True)
data_df.drop('prefer_ad', axis=1, inplace=True)
data_df.drop('prefer_overall', axis=1, inplace=True)
data_df.drop('ethnicity_race_specify', axis=1, inplace=True)
data_df.drop('gender_specify', axis=1, inplace=True)
data_df.drop('why_drink_other', axis=1, inplace=True)
data_df.drop('additions_other', axis=1, inplace=True)
data_df.drop('favorite_specify', axis=1, inplace=True)
data_df.drop('purchase_other', axis=1, inplace=True)
data_df.drop('purchase', axis=1, inplace=True)
data_df.drop('brew_other', axis=1, inplace=True)
```

```python
for column in data_df.columns:
    unique_values = data_df[column].dropna().unique()
    print(f"🔹 {column} (unique count {len(unique_values)})")
    print(unique_values)  # 앞 10개만 출력
    print("-" * 50)
```

```python
for column in data_df.columns:
    print(f"\n--- {column} ---")
    print(data_df[column].value_counts())
```

```python
multiselect_columns = ['where_drink', 'brew', 'additions', 'dairy', 'sweetener']
for col in multiselect_columns:
    dummies = data_df[col].str.get_dummies(sep=', ')
    dummies = dummies.add_prefix(f"{col}_")
    data_df = pd.concat([data_df, dummies], axis=1)
```

```python
data_df.head()
```

```python
where_drink_cols = [col for col in data_df.columns if col.startswith('where_drink_')]

where_drink_counts = data_df[where_drink_cols].sum().sort_values(ascending=False)

n_total = len(data_df)
where_drink_perc = (where_drink_counts / n_total) * 100

# 확인
print("=== 'where_drink' count ===")
print(where_drink_counts)
print("=== 'where_drink' percentage(%) ===")
print(where_drink_perc.round(2))
```

```python
plt.bar(where_drink_counts.index, where_drink_counts.values)
plt.title("Frequency of 'where_drink' Selections")
plt.xticks(rotation=90)
plt.ylabel("Number of Selections")
plt.show()
```

```python
data_df['at_home_flag'] = data_df['where_drink_At home']

ct = pd.crosstab(data_df['at_home_flag'], data_df['cups'])
print(ct)
```

```python
data_df.groupby('at_home_flag')['expertise'].mean()
```

```python
multiselect_columns = ['where_drink', 'brew', 'additions', 'dairy', 'sweetener']

for col in multiselect_columns:

    dummies_cols = [c for c in data_df.columns if c.startswith(col + "_")]

    counts = data_df[dummies_cols].sum().sort_values(ascending=False)

    print(f"=== {col} count ===")
    print(counts, "\n")

    plt.bar(counts.index, counts.values)
    plt.title(f"Frequency of '{col}' Selections")
    plt.xticks(rotation=90)
    plt.ylabel("Number of Selections")
    plt.show()
```

```python
ct_t = ct.T  # cups x at_home_flag
ct_t_norm = ct_t.div(ct_t.sum(axis=1), axis=0)

ct_t_norm.plot(kind='bar', stacked=True, figsize=(8,5))

plt.title("Coffee Cup Distribution by At-Home Flag (100% Stacked)")
plt.xlabel("Cups per Day")
plt.ylabel("Proportion")
plt.xticks(rotation=0)
plt.legend(title="At Home Flag", loc="upper right")
plt.show()
```

```python
location_cols = [col for col in data_df.columns if col.startswith('where_drink_')]
grouped = data_df.groupby('age')[location_cols].sum()
```

```python
print(grouped)
```

```python
grouped_ratio = grouped.div(grouped.sum(axis=1), axis=0)

age_order = ["<18 years old", "18-24 years old", "25-34 years old",
             "35-44 years old", "45-54 years old", "55-64 years old", ">65 years old"]

grouped = grouped.reindex(age_order)
grouped_ratio = grouped_ratio.reindex(age_order)
```

```python
grouped_ratio.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title("Proportion of Where People Drink Coffee by Age")
plt.xlabel("Age Group")
plt.ylabel("Proportion")
plt.legend(title="Where Drink", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
additions_cols = [col for col in data_df.columns if col.startswith('additions_')]
grouped_age_additions = data_df.groupby('age')[additions_cols].sum()
```

```python
grouped_age_additions_ratio = grouped_age_additions.div(grouped_age_additions.sum(axis=1), axis=0)

grouped_age_additions = grouped_age_additions.reindex(age_order)
grouped_age_additions_ratio = grouped_age_additions_ratio.reindex(age_order)
```

```python
grouped_age_additions_ratio.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title("Proportion of Coffee Additions by Age")
plt.xlabel("Age Group")
plt.ylabel("Proportion")
plt.legend(title="Additions", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
cup_order = ['Less than 1', 1, 2, 3, 4, 'More than 4']
ct_age_cup = pd.crosstab(data_df['age'], data_df['cups'])
ct_age_cup = ct_age_cup.reindex(index=age_order, columns=cup_order)
print(ct_age_cup)
```

```python
ct_age_cup_ratio = ct_age_cup.div(ct_age_cup.sum(axis=1), axis=0)

ct_age_cup_ratio.plot(kind='bar', stacked=True, figsize=(10,6))

plt.title("Proportion of Cups by Age Group (100% Stacked)")
plt.xlabel("Age Group")
plt.ylabel("Proportion")
plt.legend(title="Cups", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
